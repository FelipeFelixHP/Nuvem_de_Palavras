```python
# importar os pacotes necessários
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
# importar o arquivo csv em um df
df = pd.read_excel("Meu aniversario.xlsx")
# eliminar as colunas com valores ausentes
summary = df.dropna(subset=['Mensagens'], axis=0)['Mensagens']
#concatenar as palavras
todas_palavras = " ".join(p for p in summary)
# lista de stopword / palavra de parada
stopwords = set(STOPWORDS)
stopwords.update(["da", "meu", "em", "você", "de", "ao", "os", "Feliz aniversario", "Felipe", "Parabens", "Parabéns", "que", "é", "um", "muito", "Feliz aniversário", "Feliz", "por", "tudo", "todo",
print("Quantidade de palavras removidas: {}".format(len(stopwords)))

# endereço LOCAL da SUA imagem
niver_mask = np.array(Image.open("coracao.jpg"))
# gerar uma wordcloud
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="black",
                      width=1000, height=1000, max_words=3000,
                      mask=niver_mask, max_font_size=200,
                      min_font_size=1).generate(todas_palavras)
# mostrar a imagem final
fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
plt.imshow(wordcloud)
wordcloud.to_file("niveeeer.png")
```
